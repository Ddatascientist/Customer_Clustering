# Importing Liberaries

```python
import pandas as pd
import numpy as np
```

```python
# A function just to read files
def wrangle(filepath):

  df = pd.read_excel(filepath, parse_dates=True)

  return df
```

```python
data_1 = wrangle('/content/Data_1.xlsx')
data_2 = wrangle('/content/Data_2.xlsx')
data_3 = wrangle('/content/Data_3.xlsx')
data_4 = wrangle('/content/Data_4.xlsx')
data_5 = wrangle('/content/Data_5.xlsx')
data_6 = wrangle('/content/Data_6.xlsx')
```

```python
data_3 = data_3.rename(columns={"Product_Name": "Product Name"})
data_3.head()
```

```python
# Merging all the files together
df = pd.concat([data_1, data_2, data_3, data_4, data_5, data_6])
print(df.shape)
df.head(10)
```

```python
# rearranging the columns
df = df.iloc[:, [2, 0, 1, 3, 4, 5]]
df.head()
```

```python
df[df.duplicated(subset=['Invoice_no'])]
```

```python
# putting all the customer codes in a dict
un_names = [c for c in df['Customer'].unique().tolist()]
name_dict = {}
for c in un_names:
  code = str(np.random.randint(100, 999)) + c[:5]
  name_dict[c] = code
name_dict["Oluchi Anohu"]
```

```python
# A function to encode the customer names
def coding(x):
  return name_dict[x]
```

```python
df['Customer_ID'] = df['Customer'].apply(coding)
df.head()
```

```python
df.drop('Customer', axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df = df.iloc[:, [0, 1, 5, 2, 3, 4]]
df.head()
```

```python
# file ready and extracted for ML project
df.to_excel('Soyum_data.xlsx', index=False)
```
